1. Scrape the details of most viewed videos on YouTube from Wikipedia.


Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos


You need to find following details:


A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)


rank=[]
name=[]
artist=[]
views=[]
upload=[]

for i in driver.find_elements_by_xpath('//table[3]//tbody//tr//td[1]'):
    rank.append(i.text)
    
for i in driver.find_elements_by_xpath('//table[3]//tbody//tr//td[2]'):
    name.append(i.text)
    
for i in driver.find_elements_by_xpath('//table[3]//tbody//tr//td[3]'):
    artist.append(i.text)
    
for i in driver.find_elements_by_xpath('//table[3]//tbody//tr//td[4]'):
    views.append(i.text)
    
for i in driver.find_elements_by_xpath('//table[3]//tbody//tr//td[5]'):
    upload.append(i.text)
    
    
que1=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Views(billions)':views,'Upload Date':upload})
que1
```
