```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
#Question 1
url= 'https://www.naukri.com/'
driver.get(url)


job=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[1]/div/div/div/input')
job.send_keys('Data Analyst')

loc=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')
loc.send_keys('Bangalore')

search=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
search.click()
```

```python
job_title=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
job_titles=[]
for i in job_title:
    job_titles.append(i.text)
job_titles=job_titles[0:10]

company=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
company_name=[]
for i in company:
    company_name.append(i.text)
company_name=company_name[0:10]

location_tag=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
location=[]
for i in location_tag:
    location.append(i.text)
location=location[0:10]

exp=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]')
experience=[]
for i in exp:
    experience.append(i.text)
experience=experience[0:10]

que1=pd.DataFrame({'Job Title':job_titles,'Company Name':company_name,'Location':location,'Experience':experience})
que1
```

```python
# Question 4

url='https://www.flipkart.com/'
driver.get(url)
search=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search.send_keys('sunglasses')

click=driver.find_element_by_xpath('/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
click.click()
```

```python
page_url = [] # a list containg url's of every page
url = driver.find_elements_by_xpath("//nav[@class='yFHi8N']//a")
for i in url:
    page_url.append(i.get_attribute('href'))
    
name=[]
des=[]
price=[]
dis=[]
for i in page_url[:3]:
    driver.get(i)
    title=driver.find_elements_by_class_name("_2WkVRV")
    for i in title:
         name.append(i.text)
            
    desc=driver.find_elements_by_class_name('IRpwTa')
    for i in desc:
        des.append(i.text)
        
    amount=driver.find_elements_by_class_name('_30jeq3')
    for i in amount:
        price.append(i.text)
        
    less=driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]')
    for i in less:
        dis.append(i.text)
    
```

```python
name=name[:100]
des=des[:100]
price=price[:100]
dis=dis[:100]

que4=pd.DataFrame({'Name':name,'Description':des,'Price':price,'Discount':dis})
que4
```

```python
# Question 10

url='https://www.ambitionbox.com/'
driver.get(url)

salary=driver.find_element_by_xpath('//a[@title="Company Salaries"]')
salary.click()

search=driver.find_element_by_id('jobProfileSearchbox')
search.send_keys('Data Scientist')
```

```python

click=driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p')
click.click()
```

```python
name=[]
na=driver.find_elements_by_xpath('//div[@class="name"]')
for i in na:
    name.append(i.text.split('\n')[0])
    
total_salary=[]
ts=driver.find_elements_by_xpath('//div[@class="name"]')
for i in ts:
    total_salary.append(i.text.split('\n')[1])
    
exp=[]
ex=driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
for i in ex:
    exp.append(i.text.split('\n')[2])
    
avg_sal=[]
avg=driver.find_elements_by_xpath('//p[@class="averageCtc"]')
for i in avg:
    avg_sal.append(i.text)
    
min_sal=[]
mins=driver.find_elements_by_xpath('//div[@class="value body-medium"]')
for i in mins:
    min_sal.append(i.text)
    
min_sal=min_sal[0::2]

max_sal=[]
maxs=driver.find_elements_by_xpath('//div[@class="value body-medium"]')
for i in maxs:
    max_sal.append(i.text)
    
max_sal=max_sal[0::2]

que10=pd.DataFrame({'Company Name':name,'Total Salary':total_salary,'Experience':exp,'Average Salary':avg_sal,'Minimum Salary':min_sal,'Maximum Salary':max_sal})
que10
```
