```python
import selenium
from selenium import webdriver
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Chrome('chromedriver.exe')  #intiating the chrome
```

```python


url = "https://www.amazon.in/"

driver.get(url)

time.sleep(3)

srch_items = ['laptops','phones','headphones','smart watches', 'Professional Cameras', 'Printers', 

              'Monitors', 'Home theater', 'Router']

review=[]
rating=[]
helpful=[]
title=[]

for i in srch_items:

    srchbar = driver.find_element_by_id("twotabsearchtextbox")

    srchbar.send_keys(i)

    driver.find_element_by_id("nav-search-submit-button").click()  #clicking on search button

    time.sleep(1)
    
    url=[]
    for i in driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]'):
        url.append(i.get_attribute('href'))
        
    for i in url:
        driver.get(i)
        time.sleep(2)
    
        try:
            collect=driver.find_element_by_xpath('//a[@class="a-link-emphasis a-text-bold"]')
            collect.click()
        except: break

        url_=[]
        star=driver.find_elements_by_xpath('//tr[@class="a-histogram-row a-align-center"]//td[@class="a-span10"]//a[@class="a-link-normal"]')
        for m in star:
            url_.append(m.get_attribute('href'))

        for i in url_:
            driver.get(i)
            time.sleep(2)
            

            try:
                rev=driver.find_elements_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div/div/div/div/span/span')
                for k in rev:
                    review.append(k.text)
            except StaleElementReferenceException:
                driver.refresh()

            try:
                star=driver.find_elements_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div/div/div/div/a/i/span')
                for k in star:
                    rating.append(k.get_attribute('innerHTML'))
            except StaleElementReferenceException:
                driver.refresh()

            try:
                hf=driver.find_elements_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div/div/div/div/div/span/div[1]/span')
                for k in hf:
                    helpful.append(k.text)      
            except NoSuchElementException:
                helpful.append('Not Available')
            except StaleElementReferenceException:
                driver.refresh()

            try:
                tt=driver.find_elements_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div/div/div/div/a/span')
                for k in tt:
                    title.append(k.text)
            except StaleElementReferenceException:
                driver.refresh()

```
